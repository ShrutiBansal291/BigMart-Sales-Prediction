###***Features information***:

* **Item_Identifier** : A unique identifier for each item. It helps distinguish between different products, which can have different sales patterns.
* **Item_Weight** : The weight of each item. Heavier items might incur higher transportation or storage costs, affecting pricing and, in turn, sales.
* **Item_Fat_Content** : Indicates whether an item is low fat, regular fat, or another type. This can impact customer preference and sales for specific items.
* **Item_Visibility** : Measures the visibility of an item in the outlet. Higher visibility may lead to higher sales due to increased customer attention.
* **Item_Type** : Describes the category of the item. Different item types may have different sales trends and customer demand.
* **Item_MRP (Maximum Retail Price)** : The maximum price at which an item can be sold. Pricing has a direct impact on sales; lower prices may increase demand.
* **Outlet_Identifier** : A unique identifier for each outlet. Different outlets may have different customer bases and sales trends.
* **Outlet_Establishment_Year** : The year the outlet was established. Older outlets may have a more established customer base, impacting sales.
* **Outlet_Size** : Describes the size of the outlet (e.g., low,medium,high). Larger outlets may have higher sales due to more inventory and customer traffic.
* **Outlet_Location_Type** : Describes the type of location (e.g., urban, rural). Location type can impact the customer base and sales trends.
* **Outlet_Type** : Describes the type of outlet (e.g., supermarket, grocery store). Different outlet types may attract different types of customers and influence sales.
* **Item_Outlet_Sales**: The target variable representing the sales of a specific item at a specific outlet. The goal of the project is to predict this variable.

###  Data Splitting

````python
X = data.drop(columns='Item_Outlet_Sales', axis=1)
Y = data['Item_Outlet_Sales']
````

````python
# Split your data to train and test. Choose Splitting ratio wisely.
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
````

This helps the model generalize well to unseen data, which is crucial for real-world performance.

## *** ML Model Implementation***

### ML Model - 1

````python
#Linear Regression
from sklearn.linear_model import LinearRegression
L = LinearRegression()
````
